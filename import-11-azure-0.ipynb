Pass all tweets to the Aylien API and save the two scores returned for polarity (positive or negative sentiment scored between +1 and -1) and subjectivity (subjective or objective comments scored between +1 and -1).

A free Aylien account grants a thousand requests to their text analysis API within a fourteen day window.  Aylien also provide free access to their News API.  Several weeks before the Brexit deadline of 29th March 2019 the News API was accessed to collect text but these runs were discontinued because the text was always neutral and objective.  At least it could be used for comparison purposes against the variety expressed on Twitter.

In mid-2009 EU Parliament representatives were elected by member states.  In the UK the Conservatives were the leading political party and sent the most representatives to Brussels and UKIP surged into second place with 16% of the popular vote ... 'first past the post' system meant that Labour sent more candidates.

Error with month used up valuable twurl requests ...  meant that only the days either side of the election were collected.  Searching for #EU and #UK was successful (though #UK returned many pornographic links, deleted from the csv) but #Farage returned no tweets, other tweets showed that @Nigel_Farage would have been a better search term, hinting that it was later on in Twitter's history that users started using #Name in discussion instead of the @username handle.

```python
import time
import pandas as pd
import numpy as np
from aylienapiclient import textapi

# following website  https://www.geeksforgeeks.org/get-post-requests-using-python/

tweetDF0 = pd.read_csv('aylien703rows.csv')

azureList = list(range(0, 5))
print(azureList)

for i in azureList:
    tweetText = tweetDF0.iloc[i]['Tweets']
    tweetInQuotes = '"%s"' % tweetText                     # testing showed that Aylien Text API used double-quotes
        
    aylienClient = textapi.Client("2e3652f1", "77e0e2e019d20ea8aba8ad12572507a7")
    aylienSentiment = aylienClient.Sentiment({"text": tweetInQuotes})
    # test value # aylienSentiment = {'polarity': 'negative', 'subjectivity': 'objective', 'text': "- wert' Always been/ @labour but can't help but respect @David_Cameron putting his neck on the line for #Remain while @jeremycorbyn took back seat", 'polarity_confidence': 0.5615213513374329, 'subjectivity_confidence': 0.9998733011656746}
    
    tweetDF0.at[i, 'aylienPolarity'] = aylienSentiment['polarity_confidence']
    tweetDF0.at[i, 'aylienSubjectivity'] = aylienSentiment['subjectivity_confidence']  
    
    print("Filling in row", i, "in range 601 to 1201")
    print("Tweet sent is", tweetInQuotes)
    print('Polarity value returned is', aylienSentiment['polarity_confidence'])
    print('Subjectivity value returned is', aylienSentiment['subjectivity_confidence'])
    time.sleep(3)                                          # developer Text API limited to 2 per second, I'm cautious
    
#tweetDF0.iloc[:605] 
tweetDF0.to_csv('aylien1200rows.csv')

# earlier test results
# error received when requesting over 1000 tweet analyses "HttpError: <HttpError 429 when requesting https://api.aylien.com/api/v1/sentiment returned "Too Many Requests">"
# 
# aylienClient = textapi.Client("2e3652f1", "77e0e2e019d20ea8aba8ad12572507a7")
# aylienSentiment = aylienClient.Sentiment({"text": "- wert' Always been/ @labour but can't help but respect @David_Cameron putting his neck on the line for #Remain while @jeremycorbyn took back seat"})
# print(aylienSentiment)
# {'polarity': 'negative', 'subjectivity': 'objective', 'text': "- wert' Always been/ @labour but can't help but respect @David_Cameron putting his neck on the line for #Remain while @jeremycorbyn took back seat", 'polarity_confidence': 0.5615213513374329, 'subjectivity_confidence': 0.9998733011656746}


```
