Pass all tweets to the Aylien API and save the two scores returned for polarity (positive or negative sentiment scored between +1 and -1) and subjectivity (subjective or objective comments scored between +1 and -1).

A free Aylien account grants a thousand requests to their text analysis API within a fourteen day window.  Aylien also provide free access to their News API.
```python
import time
import pandas as pd
import numpy as np
from aylienapiclient import textapi

tweetDF0 = pd.read_csv('proAylien-0.csv')

aylienList = list(range(120, 27003))


for i in aylienList:
    tweetText = tweetDF0.iloc[i]['Tweets']
    tweetInQuotes = '"%s"' % tweetText                     # testing showed that Aylien Text API used double-quotes
        
    aylienClient = textapi.Client("2e3652f1", "ee8352156ef7eeb796d443ea9b64be1a")
    aylienSentiment = aylienClient.Sentiment({"text": tweetInQuotes})
    # test value # aylienSentiment = {'polarity': 'negative', 'subjectivity': 'objective', 'text': "- wert' Always been/ @labour but can't help but respect @David_Cameron putting his neck on the line for #Remain while @jeremycorbyn took back seat", 'polarity_confidence': 0.5615213513374329, 'subjectivity_confidence': 0.9998733011656746}
    
    tweetDF0.at[i, 'proAylienPolarity'] = aylienSentiment['polarity_confidence']
    tweetDF0.at[i, 'proAylienSubjectivity'] = aylienSentiment['subjectivity_confidence']
    time.sleep(2)
    if i % 90 == 8:                                                                        # tracking progress
        print("Row", i, "returned proAylienPolarity", aylienSentiment['polarity_confidence'])

    
tweetDF0.iloc[0:] 
#tweetDF0.to_csv('proAylien-1.csv')   keep it safe


```

The first analysis is the for the earliest vote, the European Parliament elections in June 2009.  Tweets were only retrieved for the third and fifth of June without any tweets retrieved for the fourth of June, the day of the UK elections, because the Twitter account downloads were used up.  A premium Twitter developer account allows a hundred requests in a calendar month.  Each request will return up to five hundred tweets against a search term.  In comparison to later requests in 2014 and 2016 there are far fewer tweets returned in each request using the two search terms #EU and #UKIP.  (No results at all were returned for #Farage, it may be that @Nigel_Farage was the main way he was identified in 2009.)

The amount of tweets for #EU grew slightly from the day before the elections to the day after, the size of the file downloaded grew from 240kB to 280kB, while the #UKIP file size shrank from 130kB to 60kB.  But how did the sentiment of the tweets change?

```python
###### if the message "<Figure size 640x480 with 1 Axes>" appears please run this cell code again

import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib
import matplotlib.pyplot as plt

def tweetSubset(bigDF, columnNameList):
    newDF = bigDF.loc[:, columnNameList]
    return newDF

def tweetSubsetByDate(bigDF, columnNameList, whatDate):
    allDates = bigDF.loc[:, columnNameList]
    newDF = allDates.loc[allDates['TweetDate'] == whatDate]
    return newDF

tweetDFwhole = pd.read_csv('proAylien-1.csv')

# Bar chart of simple mean before and after European Parliament elections on Thursday 4th June 2009
# other elections in other European countries were held on Wednesday 3rd and Friday 5th June
dayBeforeEUP = tweetSubsetByDate(tweetDFwhole, ['TweetDate', 'TweetTime', 'proAylienPolarity', 'Lang', 'Tweets'], '2009-06-03')
dayAfterEUP = tweetSubsetByDate(tweetDFwhole, ['TweetDate', 'TweetTime', 'proAylienPolarity', 'Lang', 'Tweets'], '2009-06-05')
crudeMeanBeforeEUP = dayBeforeEUP['proAylienPolarity'].mean()
crudeMeanAfterEUP = dayAfterEUP['proAylienPolarity'].mean()

objects = ('Wed 3rd June', 'Fri 5th June')
y_pos = np.arange(len(objects))
performance = [crudeMeanBeforeEUP, crudeMeanAfterEUP]
 
plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('0 is negative, 1 is positive')
plt.title('Mean Twitter #EU and #UKIP sentiment before and after 2009 European Parliament Elections')
plt.show()
```

```python
# Compare sentiment for #EU and #UKIP tweets in the two days

# from https://www.pythonprogramming.in/pandas-examples/how-to-find-all-rows-in-a-dataframe-that-contain-a-substring.html
# from https://www.w3resource.com/graphics/matplotlib/barchart/matplotlib-barchart-exercise-11.php

def tweetSubsetByHashtag(bigDF, whatHashtag):
    hashtagDF = bigDF[bigDF['Tweets'].str.contains(whatHashtag)]
    return hashtagDF

hashEUbeforeEUP = tweetSubsetByHashtag(dayBeforeEUP, '#EU')
hashUKIPbeforeEUP = tweetSubsetByHashtag(dayBeforeEUP, '#UKIP')
hashEUafterEUP = tweetSubsetByHashtag(dayAfterEUP, '#EU')
hashUKIPafterEUP = tweetSubsetByHashtag(dayAfterEUP, '#UKIP')
#hashEUbeforeEUP[4:15]

crudeMeanEUbeforeEUP = hashEUbeforeEUP['proAylienPolarity'].mean()
crudeMeanUKIPbeforeEUP = hashUKIPbeforeEUP['proAylienPolarity'].mean()
crudeMeanEUafterEUP = hashEUafterEUP['proAylienPolarity'].mean()
crudeMeanUKIPafterEUP = hashUKIPafterEUP['proAylienPolarity'].mean()

#print('crudeMeanUKIPafterEUP is', crudeMeanUKIPafterEUP)
#displayEU-UKIP = np.array([[crudeMeanEUbeforeEUP, crudeMeanUKIPbeforeEUP], [crudeMeanEUafterEUP, crudeMeanUKIPafterEUP]])
displayEUukip = [[crudeMeanEUbeforeEUP, crudeMeanUKIPbeforeEUP], [crudeMeanEUafterEUP, crudeMeanUKIPafterEUP]]
axesDF = DataFrame(displayEUukip, columns=['#EU','#UKIP'], index=['Before','After'])

axesDF.plot(kind = 'bar')
#plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
#plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.ylabel('0 is negative, 1 is positive')
plt.title('Change in Twitter #EU and #UKIP sentiment before and after 2009 European Parliament Elections')
plt.show()

```

So the increase in positive sentiment after the elections is only with tweets containing the #EU hashtag.
